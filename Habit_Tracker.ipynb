```python
from IPython.core.display import display, HTML

html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Habit Tracker</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            background-color: #f0f0f0;
            margin: 0;
            padding: 0;
            height: 100vh;
            display: flex;
            justify-content: center;
            align-items: center;
        }

        .container {
            background-color: #fff;
            padding: 20px;
            border-radius: 10px;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
            width: 350px;
            text-align: center;
        }

        h1 {
            margin-bottom: 20px;
            font-size: 24px;
            color: #333;
        }

        #habit-form {
            display: flex;
            flex-direction: column;
            margin-bottom: 20px;
        }

        #habit-input, #habit-notes {
            padding: 10px;
            border: 1px solid #ddd;
            border-radius: 5px;
            font-size: 16px;
            margin-bottom: 10px;
        }

        button {
            padding: 10px 15px;
            background-color: #28a745;
            color: #fff;
            border: none;
            border-radius: 5px;
            cursor: pointer;
            font-size: 16px;
        }

        button:hover {
            background-color: #218838;
        }

        .habit {
            display: flex;
            flex-direction: column;
            align-items: flex-start;
            padding: 10px;
            border: 1px solid #ddd;
            border-radius: 5px;
            margin-bottom: 10px;
            background-color: #f9f9f9;
        }

        .habit-name {
            font-size: 16px;
            text-align: left;
        }

        .progress-bar {
            width: 100%;
            height: 10px;
            background-color: #ddd;
            border-radius: 5px;
            margin: 5px 0;
        }

        .progress-bar-inner {
            height: 100%;
            background-color: #28a745;
            border-radius: 5px;
            transition: width 0.3s ease-in-out;
        }

        .complete-btn, .remove-btn {
            padding: 5px 10px;
            border: none;
            border-radius: 5px;
            cursor: pointer;
            font-size: 14px;
            margin-top: 5px;
        }

        .complete-btn {
            background-color: #007bff;
            color: #fff;
        }

        .complete-btn:hover {
            background-color: #0069d9;
        }

        .remove-btn {
            background-color: #dc3545;
            color: #fff;
            margin-top: 5px;
        }

        .remove-btn:hover {
            background-color: #c82333;
        }

        #stats {
            margin-top: 20px;
            font-size: 16px;
        }
    </style>
</head>
<body>
    <div class="container">
        <h1>Habit Tracker</h1>
        <form id="habit-form">
            <input type="text" id="habit-input" placeholder="Add a new habit" required>
            <input type="text" id="habit-notes" placeholder="Notes (optional)">
            <button type="submit">Add Habit</button>
        </form>
        <div id="habits-container"></div>
        <div id="stats"></div>
    </div>
    <script>
        const habitForm = document.getElementById('habit-form');
        const habitInput = document.getElementById('habit-input');
        const habitNotes = document.getElementById('habit-notes');
        const habitsContainer = document.getElementById('habits-container');
        const statsDiv = document.getElementById('stats');

        let habits = JSON.parse(localStorage.getItem('habits')) || [];

        function renderHabits() {
            habitsContainer.innerHTML = '';
            habits.forEach((habit, index) => {
                const habitDiv = document.createElement('div');
                habitDiv.classList.add('habit');

                const habitName = document.createElement('span');
                habitName.classList.add('habit-name');
                const habitStartDate = new Date(habit.startDate);
                const today = new Date();
                const daysTracked = Math.floor((today - habitStartDate) / (1000 * 60 * 60 * 24));
                habitName.textContent = `${habit.name} (${daysTracked} days)${habit.notes ? ': ' + habit.notes : ''}`;

                const progressBar = document.createElement('div');
                progressBar.classList.add('progress-bar');

                const progressBarInner = document.createElement('div');
                progressBarInner.classList.add('progress-bar-inner');
                progressBarInner.style.width = `${habit.progress}%`;
                progressBar.appendChild(progressBarInner);

                const completeBtn = document.createElement('button');
                completeBtn.classList.add('complete-btn');
                completeBtn.textContent = 'Complete';
                completeBtn.onclick = () => completeHabit(index);

                const removeBtn = document.createElement('button');
                removeBtn.classList.add('remove-btn');
                removeBtn.textContent = 'Remove';
                removeBtn.onclick = () => removeHabit(index);

                habitDiv.appendChild(habitName);
                habitDiv.appendChild(progressBar);
                habitDiv.appendChild(completeBtn);
                habitDiv.appendChild(removeBtn);

                habitsContainer.appendChild(habitDiv);
            });
            displayStats();
        }

        function completeHabit(index) {
            if (habits[index].progress < 100) {
                habits[index].progress += 10;
                if (habits[index].progress > 100) habits[index].progress = 100;
                localStorage.setItem('habits', JSON.stringify(habits));
                renderHabits();
            }
        }

        function removeHabit(index) {
            habits.splice(index, 1);
            localStorage.setItem('habits', JSON.stringify(habits));
            renderHabits();
        }

        function displayStats() {
            const totalHabits = habits.length;
            const averageProgress = habits.reduce((sum, habit) => sum + habit.progress, 0) / totalHabits;

            statsDiv.innerHTML = `Total Habits: ${totalHabits}<br>Average Progress: ${averageProgress.toFixed(2)}%`;
        }

        function setReminder(habit) {
            // Example reminder setup (not functional for actual reminders)
            const reminderTime = 24 * 60 * 60 * 1000; // 1 day
            setTimeout(() => {
                alert(`Reminder: Time to work on your habit: ${habit.name}`);
            }, reminderTime);
        }

        habitForm.onsubmit = function(e) {
            e.preventDefault();
            const newHabit = {
                name: habitInput.value,
                notes: habitNotes.value,
                progress: 0,
                startDate: new Date().toISOString().split('T')[0] // ISO date format (YYYY-MM-DD)
            };
            habits.push(newHabit);
            habitInput.value = '';
            habitNotes.value = '';
            localStorage.setItem('habits', JSON.stringify(habits));
            renderHabits();
            setReminder(newHabit);
        };

        renderHabits();
    </script>
</body>
</html>
"""

# Display the HTML content
display(HTML(html_content))
```
